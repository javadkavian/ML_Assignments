```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

# Generate two 2D datasets (replace this with your own datasets if needed)
X1 = np.array([[4, 1], [2, 4], [2, 3], [3, 6], [4, 4]])
X2 = np.array([[9, 10], [6, 8], [9, 5], [8, 7], [10, 8]])

# Combine the datasets and create labels
X = np.vstack((X1, X2))
y = np.array([0]*5 + [1]*5)

# Apply LDA
lda = LDA(n_components=1)
X_lda = lda.fit_transform(X, y)

# Find the greatest eigenvalue
eigenvalues = lda.explained_variance_ratio_
greatest_eigenvalue = np.max(eigenvalues)

print(f"The greatest eigenvalue is: {greatest_eigenvalue}")

# Plot the original datasets
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.scatter(X1[:, 0], X1[:, 1], label='Class 1')
plt.scatter(X2[:, 0], X2[:, 1], label='Class 2')
plt.title("Original Datasets")
plt.legend()

# Plot the LDA transformed dataset
plt.subplot(1, 2, 2)
plt.scatter(X_lda[y == 0], np.zeros_like(X_lda[y == 0]), label='Class 1')
plt.scatter(X_lda[y == 1], np.zeros_like(X_lda[y == 1]), label='Class 2')
plt.title("LDA Transformed Dataset")
plt.legend()

plt.tight_layout()
plt.show()
```
