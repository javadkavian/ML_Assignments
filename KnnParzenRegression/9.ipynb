```python
df.iloc[:, 0:-1] = df.iloc[:, 0:-1].apply(lambda x : ((x - x.min())/(x.max() - x.min())))
df.head()
```

```python
def myTrainTestSplit(X, Y, test_size=0.2, random_state=42):
    if random_state is not None:
        np.random.seed(random_state)
    indices = np.arange(X.shape[0])
    np.random.shuffle(indices)
    test_length = int(len(X) * test_size)
    test_indices = indices[:test_length]
    train_indices = indices[test_length:]
    X_train = X[train_indices]
    X_test = X[test_indices]
    Y_train = Y[train_indices]
    Y_test = Y[test_indices]
    
    return X_train, X_test, Y_train, Y_test
```

```python
X = df.iloc[:, 0:-1].to_numpy()
Y = df.iloc[:, -1].to_numpy()
X_train, X_test, Y_train, Y_test = myTrainTestSplit(X, Y)
```

```python
cols = df.columns.tolist()
featureAccess = {cols[i]:i for i in range(len(cols))}
```

```python
selectedFeature = 'MntCoffee'
x_train = X_train[:, featureAccess[selectedFeature]].reshape((X_train.shape[0], 1))
x_test = X_test[:, featureAccess[selectedFeature]].reshape((X_test.shape[0], 1))
Y_train = Y_train.reshape((Y_train.shape[0], 1))
Y_test = Y_test.reshape((Y_test.shape[0], 1))
N = x_train.shape[0]
```

```python
y_sum = np.sum(Y_train)
x_sum = np.sum(x_train)
a = (np.dot(x_train.flatten(), (2*Y_train - (2*y_sum/N)).flatten())) / (np.dot(x_train.flatten(), (2*x_train - (2*x_sum)/N).flatten()))
b = np.sum(Y_train - a*x_train) / N
```

```python
def linear_regressor(query):
    return a*query + b
```

```python
x = np.arange(0, 1500, 0.1)
x = ((x - np.min(x))/(np.max(x) - np.min(x)))
y = linear_regressor(x)
plt.figure()
plt.scatter(df[selectedFeature], df['NumPurchases'], alpha=0.5, color='blue')
plt.scatter(x, y, alpha=0.1, color='red')
plt.title("Fit line")
plt.grid()
```

```python
def calculate_rmse(yMain, yPrediction):
    yMain = np.array(yMain)
    yPrediction = np.array(yPrediction)
    mse = np.mean((yMain - yPrediction) ** 2)
    rmse = np.sqrt(mse)
    return rmse


def calculate_r2(yMain, yPrediction):
    mean_yMain = sum(yMain) / len(yMain)
    
    tss = sum((y - mean_yMain) ** 2 for y in yMain)
    
    rss = sum((yMain[i] - yPrediction[i]) ** 2 for i in range(len(yMain)))
    
    r2 = 1 - (rss / tss)
    
    return r2
```

```python
y_pred = []
for test in x_test:
    y_pred.append(linear_regressor(test))

print("RootMeanSquareError is : ", calculate_rmse(Y_test.tolist(), y_pred))
print("R2 score is : ", calculate_r2(Y_test, y_pred)[0])
```

```python
class MyLinearRegression:
    def __init__(self, iter = 1000, learning_rate = 0.0001):
        self.learningRate = learning_rate
        self.numberOfIterations = iter
        self.theta = None
    def fit(self, x, y):
        x = np.hstack((np.ones((x.shape[0], 1)), x))
        self.theta = np.zeros((x.shape[1], 1))

        for i in range(self.numberOfIterations):
            self.theta = self.theta + self.learningRate*(((y - np.matmul(x, self.theta))*x).sum(axis=0).reshape(x.shape[1], 1))

    def pred(self, x):
        x = np.hstack((np.ones((1, )), x))
        return np.dot(x, self.theta)
        

```

```python
x_train = X_train[:, [featureAccess['Marital_Status'], featureAccess['UsedCampaignOffer'], featureAccess['MntCoffee']]].reshape((X_train.shape[0], 3))
x_test = X_test[:, [featureAccess['Marital_Status'], featureAccess['Teenhome'], featureAccess['MntCoffee']]].reshape((X_test.shape[0], 3))
Y_train = Y_train.reshape((Y_train.shape[0], 1))
```

```python
linearModel = MyLinearRegression()
linearModel.fit(x_train, Y_train)
y_pred = []
for test in x_test:
    y_pred.append(linearModel.pred(test))    
```

```python
print("RootMeanSquareError is : ", calculate_rmse(Y_test.tolist(), y_pred))
print("R2 score is : ", calculate_r2(Y_test, y_pred)[0])
```
