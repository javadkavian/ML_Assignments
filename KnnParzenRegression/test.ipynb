```python
import numpy as np
from scipy.spatial import Voronoi, voronoi_plot_2d
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

# Define your points for Class A and Class B
points_A = np.array([[0, 0], [0, 1], [1, 0]])  # Replace with your coordinates for Class A
points_B = np.array([[1, 1], [1, 2], [2, 1]])  # Replace with your coordinates for Class B

# Combine the points from both classes
points = np.concatenate((points_A, points_B))

# Compute the Voronoi diagram
vor = Voronoi(points)

# Create a color map for the two classes
colors = ['red', 'blue']

# Plot the Voronoi diagram
fig, ax = plt.subplots()
voronoi_plot_2d(vor, ax=ax, show_vertices=False, line_colors='gray', line_width=2, line_alpha=0.6)

# Color the Voronoi regions
for region, point in zip(vor.regions, vor.point_region):
    if not -1 in region:
        polygon = [vor.vertices[i] for i in region]
        ax.fill(*zip(*polygon), colors[points.tolist().index(points[point].tolist()) % len(colors)])

# Mark the points for Class A
ax.plot(points_A[:, 0], points_A[:, 1], 'ro', label='Class A')

# Mark the points for Class B
ax.plot(points_B[:, 0], points_B[:, 1], 'bo', label='Class B')

# Add a legend
ax.legend()

# Display the plot
plt.show()
```
