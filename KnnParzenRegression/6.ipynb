```python
import numpy as np
import matplotlib.pyplot as plt
low = 0.0 
high = 2.0
numberOfSamples = 300
Vn = 0.05
samples = np.random.uniform(low, high, numberOfSamples)
```

```python
def gaussianPdf(mean, var, x):
    return (np.exp(-((x - mean) ** 2) / (2 * var))) / (np.sqrt(2 * np.pi * var))
```

```python
x = np.arange(-100, 100, 0.1)
y = (1/numberOfSamples)*np.sum(np.array([gaussianPdf(t, Vn, x) for t in samples]).T, axis=1)
plt.plot(x, y)
plt.xlim([-5, 5])
```
